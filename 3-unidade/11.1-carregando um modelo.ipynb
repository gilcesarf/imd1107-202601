```python
import numpy as np

import sys
sys.path.insert(0, '/content/drive/MyDrive/IMD1107 - Processamento de Linguagem Natural/3-unidade/')

from toolbox import MyModel
```

### **1. Importando os dados de teste:**

```python
X_test = np.load("/content/drive/MyDrive/IMD1107 - Processamento de Linguagem Natural/3-unidade/dataset/x_test.npy")
y_test = np.load("/content/drive/MyDrive/IMD1107 - Processamento de Linguagem Natural/3-unidade/dataset/y_test.npy")

X_test.shape, y_test.shape
```

### **2. Impotando o modelo:**

```python
model = MyModel()

model.load_weights("/content/drive/MyDrive/IMD1107 - Processamento de Linguagem Natural/3-unidade/weights/model.weights.h5")

model.summary()
```

### **3. Avaliando o modelo:**

```python
model.compile(loss='mse', metrics=['mae'])

mse, mae = model.evaluate(x=X_test, y=y_test, batch_size=16)

print(f"\nMSE = {mse:.3f}, MAE = {mae:.3f}\n")
```

```python
y_pred = model.predict(X_test)

print("\ny-pred | y-test")
print("---------------")

for p, y in zip(y_pred[:10], y_test[:10]):

    print(f" {p[0]:.2f}  |  {y:.2f}")
```
