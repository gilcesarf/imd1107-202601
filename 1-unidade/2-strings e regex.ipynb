## __1. *strings* em Python:__

A linguagem oferece diversos métodos nativos para manipulação eficiente de _strings_, muitos dos quais são essenciais para o processamento e estruturação de dados textuais

```python
frase1 = "O que veio primeiro, o código ou o bug?"
print(frase1, "\n")

frase2 = "Me arruma um copo d'água, por favor."
print(frase2, "\n")

frase3 = 'O curso de PLN vai ser "tranquilo".'
print(frase3)
```

```python
def Media(x):

    """Calcula a média aritmética de uma lista de números.

    Args:
        x (list): Lista contendo valores numéricos (int ou float).

    Returns:
        float: Média dos valores da lista.

    Examples:
        >>> Media([10, 20, 30])
        20.0
        >>> Media([1.5, 2.5, 3.5])
        2.5
    """

    return sum(x)/len(x)

print(Media.__doc__)
```

```python
x = [1, 4, 6, 2, 8, 3, 9]

print(f"Média de x = {Media(x):.2f}")
```

#### __1.3.1 Split:__
<font size=3>

$\texttt{str.split(sep=None, maxsplit=-1)}$

Retorna uma lista das palavras na *string*, usando `sep` como delimitador. Se `maxsplit` for fornecido, este será a quantidade máxima de divisões. Se `maxsplit` não for especificado ou -1, não há limite para o número de divisões (todas as divisões possíveis são feitas).

```python
print(str.split.__doc__) # = help(str.split)
```

```python
'''Se sep for fornecido, delimitadores consecutivos não são agrupados
e são considerados como delimitadores de strings vazias'''

frase = "bug???"

frase_split = frase.split('?')

print(frase_split)
```

```python
# O argumento sep pode ter vários caracteres

frase = "Cara, o código... deu bug de novo!"

frase_split = frase.split('...')

print(frase_split)
print(type(frase_split))
```

<font size=3>
    
Se o argumento `sep` não for especificado ou for `None`, espaços em branco consecutivos serão tratados como um único separador. Além disso, a string resultante não conterá elementos vazios no início ou no final, mesmo que a string original possua espaços em branco à esquerda ou à direita. Como consequência, dividir uma string vazia ou composta apenas por espaços em branco com `sep=None` retornará uma lista vazia.

```python
frase = " O curso de PLN vai ser tranquilo. "

frase_split = frase.split()
print(frase_split, "\n")

frase_split = frase.split(' ')
print(frase_split, "\n")

frase_split = frase.split(' ', maxsplit=1)
print(frase_split, "\n")
```

#### __1.3.2 Count:__
<font size=3>
    
$\texttt{str.count(value, start, end)}$

Este método retorna a quantidade de vezes que determinado valor aparece na *string*.

`value:` substring a ser contada (argumento necessário).

`start:` é o índice inicial dentro da *string* onde começa a pesquisa pelo valor especificado (opcional).

`end:` é o índice final dentro da *string* onde termina a pesquisa pelo valor especificado (opcional).


```python
frase = "Me arruma um copo d'água, com um pedaço de doce de goiaba, por favor"

print("Número de letras 'a':", frase.count("a"), "\n")
print("Quantas vezes aparece 'um':", frase.count("um", 5, 35))
```

#### __1.3.3 Lower:__
<font size=3>

$\texttt{str.lower()}$

Retorna uma cópia da string com todos os caracteres minúsculos.

```python
frase = "Tu sabe o que é PLN? Pln é a área ...."
frase = frase.lower()

print(frase, "\n")
print("Quantas vezes aparece 'PLN':", frase.count("pln"))
```

#### __1.3.4 Find:__
<font size=3>
    
$\texttt{str.find(value, start, end)}$

Retorna o menor índice inicial de uma _substring_ específica em uma _string_. Se a _substring_ não for encontrada, -1 é retornado.

- `value:` _substring_ a ser procurada (argumento necessário).

- `start:` é o índice inicial dentro da _string_, onde começa a pesquisa pelo valor especificado (opcional).

- `end:` é o índice final dentro da _string_, onde termina a pesquisa pelo valor especificado (opcional).

O método $\texttt{rfind()}$ é semelhante a $\texttt{find()}$, exceto que retorna o maior índice inicial da _substring_.

```python
print("Frase:", frase, "\n")

i = frase.find("pln")
j = frase.rfind("pln")

k = frase.find("pln", 18, 25)

print(f'A string "pln" é encontrada primeiro no índice = {i}, que leva o caractere "{frase[i]}" na frase.', "\n")
print(f'A string "pln" é encontrada por último no índice = {j}, que leva o caractere "{frase[j]}" na frase.', "\n")
print(f'Entre a posição 18 e 25, encontramos "pln" no índice = {k} na frase.')
```

<font size=3>

O método $\texttt{find()}$ só é relevante quando temos interesse em saber o índice inicial de determinado elemento em uma string. \
Se o índice não é importante, utilizamos uma estrutura **condicional**.

```python
"pln" in frase
```

```python
if "pln" in frase:
    print("True")

else:
    print("False")
```

#### __1.3.5 Strip:__
<font size=3>

$\texttt{str.strip(chars)}$

Retorna uma cópia da *string* com os caracteres iniciais e finais removidos. O argumento `chars` é uma *string* que especifica o conjunto de caracteres a ser removido. Se omitido ou `None`, o argumento `chars` assume como padrão a remoção de espaços em branco. Note que `chars`não é um prefixo ou sufixo; em vez disso, todas as combinações de seus valores são removidas.

```python
print('   muito espaço   '.strip(), "\n")

print('www.example.com'.strip('cmowz.'), "\n")
```

<font size=3>

Os valores externos do argumento `chars` à esquerda e à direita são retirados da *string*. Os caracteres são removidos da extremidade inicial até chegar a um caractere de *string* que não está contido no conjunto de caracteres em `chars`. Uma ação semelhante ocorre na extremidade final.

```python
frase = '#....... Seção #4.2 .......'

print(frase.strip('.#! '))
```

#### __1.3.7 Replace:__
<font size=3>
    
$\texttt{str.replace(old, new, count)}$

- `old` = ocorrências a serem substituídas (necessário).
- `new` = pelo que as ocorrências anteriores serão substituídas (necessário).
- `count` = quantidade de ocorrências a serem substituídas (opcional).

Retorna uma cópia da *string* com todas as ocorrências da substring antiga substituídas pelas novas. Se `count` for fornecido, apenas as `count` primeiras ocorrências serão substituídas.

```python
frase = "Cara, o código... deu bug de novo!"

print(frase.replace("...", " -"), "\n")

print(f'casa -> {"casa".replace("as", "op")}', "\n")
```

## **2. Expressões Regulares (RegEx):**
<font size=3>
    
[Expressões regulares](https://pt.wikipedia.org/wiki/Express%C3%A3o_regular) (também chamadas de **regex** ou **padrões regex**) são uma linguagem concisa e poderosa para **definir regras de busca em textos**. Elas permitem localizar, extrair e manipular trechos de *strings* que seguem um determinado formato, sendo amplamente utilizadas em tarefas de **processamento de linguagem natural (PLN)**, análise de dados e validação de entradas.

<font size=3>

No Python, expressões regulares estão disponíveis por meio do módulo **`re`**, que oferece funções para pesquisar, substituir e dividir textos com base em padrões.

Um detalhe importante é o uso da **barra invertida (`\`)**, que pode ter dois papéis:

* indicar **sequências especiais** (por exemplo, `\d` para dígitos, `\s` para espaços em branco),
* ou **escapar caracteres** que têm significado especial na sintaxe (por exemplo, `\.` para representar literalmente o ponto final).

Como o próprio Python também interpreta a barra invertida em *strings* comuns (por exemplo, `\n` como quebra de linha), recomenda-se o uso de ***raw strings*** (prefixadas com `r`). Assim:

* `r"\n"` → representa literalmente os caracteres `\` e `n`,
* `"\n"` → representa uma quebra de linha.

Dessa forma, evitamos ambiguidades e tornamos os padrões mais legíveis.

**Metacaracteres mais usados:**

| Símbolo | Significado | Exemplo |
|---------|-------------|---------|
| `.`     | Qualquer caractere (exceto nova linha) | `a.c` encontra `abc`, `a1c` |
| `\d`    | Dígito (0-9) | `\d+` encontra `123` |
| `\w`    | Caractere alfanumérico | `\w+` encontra `palavra123` |
| `\s`    | Espaço em branco | `\s+` encontra espaços múltiplos |
| `^`     | Início da string | `^Olá` encontra apenas se o texto começar com "Olá" |
| `$`     | Fim da string | `fim$` encontra apenas se terminar com "fim" |
| `[]`    | Conjunto de caracteres | `[aeiou]` encontra vogais |
| `|`     | OU | `gato|cachorro` encontra qualquer um |
| `*`     | Zero ou mais repetições | `a*` encontra `"", "a", "aa"` |
| `+`     | Uma ou mais repetições | `a+` encontra `"a", "aa", "aaa"` |
| `?`     | Zero ou uma repetição | `a?` encontra `""` ou `"a"` |

[RegEx cheatlist](https://regexr.com/).

### __2.1 Sub:__
<font size=3>
    
Depois de reconhecer padrões, uma tarefa comum é substituí-los. `re.sub` troca todas as ocorrências (ou um número limitado) por um texto novo, podendo usar grupos capturados.
    
$\texttt{re.sub(pattern, repl, string)}$

Retorna uma nova _string_ com a substituição desejada.

- `pattern`: padrão para substituição.
- `repl` (de "replacement"): pelo que o padrão anterior será substituído.
- `string`: variável string a ser modificada.


```python
import re

texto = "O ano é 2023, e o Python 3.10 foi lançado em 05/10/2021."

texto_limpo = re.sub(r'\d+', '', texto)
texto_limpo
```

<font size=3>

- **`r'\d+'`**: Busca **um ou mais dígitos** (`\d` = dígito, `+` = repetição).  
- **`''`**: Substitui os números por **nada** (string vazia).  


```python
texto = "O PLN (Processamento de LINGUAGEM Natural) é INCrível."

re.sub(r'[A-ZÁ-Ú]{2,}', '[MAIÚSCULA]', texto)
```

<fotn size=3>

- **`r'[A-ZÁ-Ú]{2,}'`**: Busca 2 ou mais letras maiúsculas (incluindo acentuadas).  


```python
texto = "Olá! Como você está? Isso é um exemplo: com pontuação..."

re.sub(r'[^\w\s]', '', texto)
```

<font size=3>

- **Padrão `r'[^\w\s]'`**:
  - `\w` = qualquer caractere alfanumérico (letras, números) e underline (`_`).
  - `\s` = espaços em branco (espaço, tab, quebra de linha).
  - `[^...]` = **negação** do conjunto (ou seja, "tudo que **não** for `\w` ou `\s`").
- **Substituição**: String vazia (`''`), ou seja, remove os caracteres indesejados.
  

```python
texto = "Cadastro: 123.456.789-10, 987.654.321-00. Confirme seu CPF!"
re.sub(r'\d{3}\.\d{3}\.\d{3}-\d{2}', '***CPF***', texto)
```

<font size=3>

- **Padrão `r'\d{3}\.\d{3}\.\d{3}-\d{2}'`**:
  - `\d{3}` = três dígitos.
  - `\.` = ponto literal (escapado).
  - `-` = hífen literal.
  - Combinação típica de um CPF (`XXX.XXX.XXX-XX`).
- **Substituição**: `***CPF***` (máscara para ocultar o CPF).


```python
texto = "Texto   com    espaços    excessivos.  PLN   é   incrível!"
re.sub(r'\s+', ' ', texto)
```

<font size=3>
    
- **Padrão `r'\s+'`**:
  - `\s` = qualquer espaço em branco (inclui tab, quebra de linha).
  - `+` = uma ou mais ocorrências.
- **Substituição**: `' '` (um único espaço).


### **__2.2 Match:__**
<font size=3>

$\texttt{re.match()}$

Verifica se o padrão regex corresponde ao início da *string*.

```python
texto = "PLN é útil para análise de textos"

resultado = re.match(r"PLN", texto)

if resultado:
    print("Padrão encontrado no início!")
else:
    print("Padrão não encontrado.")
```

### **__2.3 Search:__**
<font size=3>

$\texttt{re.search()}$

Busca o padrão regex em qualquer posição da *string* (não apenas no início).

```python
texto = "A linguagem Python é poderosa para PLN."

resultado = re.search(r"Python", texto)

if resultado:
    print(f"Encontrado na posição {resultado.start()} a {resultado.end()}.")
```

### ****2.4 Fullmatch:****
<font size=3>
    
$\texttt{re.fullmatch()}$

Verifica se **toda a *string*** corresponde ao padrão regex (do início ao fim) — se houver qualquer caractere extra fora do padrão, o resultado é `None`.

É especialmente útil em **validações**, como e-mails, CPFs ou códigos de matrícula, porque garante que não haja caracteres extras fora do padrão definido.

```python
emails = ["aluno@universidade.br",
          "professor.ufrn.edu",     # inválido (falta @)
          "teste@dominio",          # inválido (sem sufixo)
          "nome.sobrenome@ufrn.br"]

padrao = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"

for e in emails:
    if re.fullmatch(padrao, e):
        print(f"{e} → válido!")
    else:
        print(f"{e} → inválido!")
```

### **__2.5 Findall:__**
<font size=3>

$\texttt{re.findall()}$

Retorna todas as ocorrências não sobrepostas do padrão como uma lista de *strings*.

```python
texto = "Os anos são 2020, 2021, 2022 e 2023."

re.findall(r"\d{4}", texto)
```

### **__2.6 Compile:__**
<font size=3>

Se o mesmo padrão será usado repetidas vezes (ex., dentro de um laço), `re.compile` melhora a legibilidade e pode trazer ganhos de desempenho.

$\texttt{re.compile()}$

Divide a *string* usando o padrão regex como delimitador.

```python
# palavras com exatamente 4 letras:
padrao = re.compile(r"\b\w{4}\b")

texto = "Isso é um teste de palavras como casa e código."

padrao.findall(texto)
```

<font size=3>

Em `\b\w{4}\b` nós temos:

- **`\b`** → **"Fronteira de palavra"** (posição onde um caractere de palavra (`\w`) é seguido ou precedido por um não-palavra, como espaço ou pontuação).  
- **`\w{4}`** → **"Exatamente 4 caracteres de palavra"** (letras, números ou `_`).  
- **`\b`** → Outra fronteira de palavra para garantir que seja uma palavra completa.
