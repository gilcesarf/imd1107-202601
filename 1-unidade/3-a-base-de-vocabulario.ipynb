# **A Base de Vocabulário**
<font size=3>
    
Assim como nós (seres humanos) necessitamos desenvolver um vocabulário básico para nos comunicarmos em diferentes idiomas, as máquinas também dependem da construção de um repertório lexical fundamental. Esse vocabulário computacional é composto não apenas por palavras isoladas, mas também por elementos linguísticos como morfemas (prefixos e sufixos), símbolos numéricos, sinais de pontuação e outros **_tokens_** textuais relevantes para a representação da linguagem.

Nesta aula, iniciaremos o processo de transformação de um conjunto de dados (frases) em representações numéricas &mdash; o __processamento de dados__ em PLN.

## __1. Tokenização:__
<font size=3>

Em PLN, os __conjuntos de dados__ (_datasets_) são tipicamente compostos por coleções de frases ou textos extensos. Quando esses conjuntos consistem em um grande volume de obras organizadas &mdash; como livros, artigos científicos ou registros linguísticos &mdash; denominamos esse compilado de **_corpus_** textual. Esses _corpora_ (no plural) são fundamentais para o treinamento de modelos de Aprendizado Profundo, pois fornecem a base linguística necessária para que os algoritmos aprendam padrões complexos da linguagem humana.  

<font size=3>

Para ilustrar a criação de uma base de vocabulário, vamos considerar as __regras de boas práticas de programação__ do [Zen of Python](https://peps.python.org/pep-0020/) como no nosso _dataset_:

````python
dataset = ["Beautiful is better than ugly",
           "Explicit is better than implicit",
           "Simple is better than complex",
           "Complex is better than complicated",
           "Flat is better than nested",
           "Sparse is better than dense",
           "Readability counts",
           "Special cases aren't special enough to break the rules",
           "Although practicality beats purity",
           "Errors should never pass silently",
           "Unless explicitly silenced",
           "In the face of ambiguity, refuse the temptation to guess",
           "There should be one - and preferably only one - obvious way to do it",
           "Although that way may not be obvious at first unless you're Dutch",
           "Now is better than never",
           "Although never is often better than right now",
           "If the implementation is hard to explain, it's a bad idea",
           "If the implementation is easy to explain, it may be a good idea",
           "Namespaces are one honking great idea - let's do more of those!"]
````

<font size=3>

Para vetorizar (transformação numérica) as 19 frases do nosso _dataset_, precisamos primeiro definir uma estratégia de segmentação (em caracteres, palavras ou subpalavras) que permita associá-las a representações numéricas.

<font size=3>

Essas unidades textuais são chamadas de __tokens__. Assim, vamos utilizar uma frase para explorar as três formas principais de **tokenização**:

   ```python
    sentence = ["Beautiful is better than ugly"]
             
   ```

- __A nível de caractere__:

   ```python
    tokens = ['B', 'e', 'a', 'u', 't', 'i', 'f', 'u', 'l', 'i', 's',
              'b', 'e', 't', 't', 'e', 'r', 't', 'h', 'a', 'n',
              'u', 'g', 'l', 'y']
             
   ```

- __A nível de palavra__:

   ```python
    tokens = ['Beautiful', 'is', 'better', 'than', 'ugly']
             
   ```
<font size=3>

- __A nível de *N-gram*__:

   ```python
    # 1-grams/unigram
    tokens = ['Beautiful', 'is', 'better', 'than', 'ugly']

    # 2-grams/bigram
    tokens = ["Beautiful is", "is better", "better than", "than ugly"]
     
   ```

<br>

[*N-gram*](https://en.wikipedia.org/wiki/N-gram) é utilizada para correlacionar a sequência de $n$ palavras de modo a extrair significados encadeados, _e.g._ ["machine learning"] é diferente de ["machine", "learning"].

A estratégia de tokenização irá __depender do tipo de tarefa__ que iremos resolver. A seguir, vamos considerar uma tokenização a nível de palavra:

````python
# Tokenização por palavras (utilizando list comprehension):

token_list = [text.split() for text in dataset]

token_list[:5] # 5 primeiras frases do dataset
````

<font size=3>

Usando a biblioteca *Natural Language Toolkit* ($\texttt{nltk}$), podemos fazer a tokenização da seguinte forma:

````python
import nltk
from nltk.tokenize import word_tokenize
````

````python
# Download do modelo tokenizador punkt_tab (necessário uma única vez):
nltk.download('punkt_tab')

text = dataset[0]

tokens = word_tokenize(text)

print("Frase:", text)
print("Lista de tokens:", tokens)
````

## __2. Normalização:__
<font size=3>

A fase de **normalização** consiste em padronizar os textos para reduzir variações irrelevantes, de modo a facilitar o processamento computacional.

<font size=3>

Por exemplos, podemos encontrar representações da mesma palavra em uma lista de tokens, como

```python
    ["The", "the", ...] ,
    
```
ou no caso,

```python
    [..., "better", "better!", ...] .
```

Fazer ou não as seguintes técnicas irá **depender da tarefa** &mdash; remoção de pontuação pode não afetar uma análise de sentimentos, mas pode afetar o processamento de um _chatbot_.

### __2.1 Conversão para minúsculo:__

````python
# Considerando a última frase:
text = token_list[-1]

print("Old:", text)

new_text = []

for token in text:

    token_processed = token.lower()

    new_text.append(token_processed)

print("New:", new_text)

text = new_text
````

<font size=3>
<br>
    
**Observação:** em algumas etapas de normalização — como *conversão para minúsculo* e *remoção de pontuação* — os textos não precisam passar pela tokenização primeiro!

````python
frase = dataset[-1]

print("Old frase:", frase)

frase = frase.lower()

print("New frase:", frase)
````

### __2.2 Remoção de pontuação:__

````python
import string

string.punctuation
````

````python
# Considerando a última frase:
print("Old:", text)

new_text = []

for token in text:

    token_processed = token.translate(str.maketrans('', '', string.punctuation))

    new_text.append(token_processed)

print("New:", new_text)
````

````python
# Considerando a última frase:
print("Old:", text)

new_text = []

for token in text:

    token_processed = token.translate(str.maketrans('', '', string.punctuation))

    if token_processed == '': # removendo '' da lista!
        continue

    new_text.append(token_processed)

print("New:", new_text)

text = new_text
````

<font size=3>
<br>

Usando **expressões regulares** sem passar pela tokenização — vamos comparar os resultados.

````python
import re
````

````python
print("Old frase:", frase)

new_frase = re.sub(r'[^\w]', ' ', frase)

print("New frase:", new_frase)
````

````python
# removendo o espaço em excesso:

print("Newer frase:", re.sub(r'\s+', ' ', new_frase))
````

````python
# removendo um apostrofe seguido de caractere alfanumérico:

print("Newer frase:", re.sub(r"'(?=\w-)", '', frase) )

# removendo um apóstrofe seguido de caractere alfanumérico OU hífen:

print("Newer frase:", re.sub(r"'(?=\w)|-", '', frase) )
````

### __2.3 Stemização:__
<font size=3>

A **Stemização** (*stemming*) remove afixos (prefixos e sufixos) das palavras, retornando um "radical" (_stem_), que nem sempre é uma palavra válida. Como em:
 - "correndo" $\rightarrow$ "corr";  
 - "felicidade" $\rightarrow$ "felic";
   
 - "A inteligência artificial ajuda médicos a diagnosticar doenças complexas." $\rightarrow$ "A inteligen artific ajud medic a diagnostic doenç complex."


Usamos essa técnica quando a tarefa prioriza a **velocidade** de processamento (_e.g._ indexação de documentos), ou quando apresenta idiomas com **poucos recursos disponíveis**.

````python
from nltk.stem import PorterStemmer

# Baixa os recursos necessários:
nltk.download('wordnet')
nltk.download('omw-1.4')

# Inicializando stemizador:
stemmer = PorterStemmer()

print("Old:", text)

new_text = []

for token in text:
    token_processed = stemmer.stem(token)

    new_text.append(token_processed)

print("New:", new_text)
````

### __2.4 Lematização:__
<font size=3>

A **Lematização** (*lemmatization*) reduz a palavra à sua forma canônica (**lema**), baseada no dicionário e análise morfológica.  Por exemplo:  
 - "correndo" $\rightarrow$ "correr" (verbo no infinitivo).  

 - "gatos" $\rightarrow$ "gato" (singular).

 - "Nós fomos ao laboratório e vimos computadores velozes." $\rightarrow$ "Nós ir a o laboratório e ver computador veloz."

Não usamos essa técnica quando a tarefa exige **precisão** (_e.g._ chatbots, análise de sentimentos), ou quando a forma da **palavra base importa** (_e.g._ sistemas de perguntas e respostas).

````python
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

# Inicializa o lematizador:
lemmatizer = WordNetLemmatizer()

print("Old:", text)

new_text = []

for token in text:
    token_processed = lemmatizer.lemmatize(token)

    new_text.append(token_processed)

print("New:", new_text)

text = new_text
````

### __2.5 Remoção de stop words:__
<font size=3>

**Stop words** são palavras extremamente comuns em um idioma que, embora importantes para a estrutura gramatical, geralmente **não carregam significado relevante** para tarefas.\
Exemplos incluem "o", "e", "de", "para" (em português) ou "the", "and", "of" (em inglês).

````python
from nltk.corpus import stopwords

# Faz o download da lista de stop words:
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

print("Algumas stopword do inglês:", list(stop_words)[:10])
````

````python
print("Old:", text)

new_text = []

for token in text:
    if token not in stop_words:
        new_text.append(token)

print("New:", new_text)
````

## __3. O dicionário de vocabulário:__
<font size=3>

Para criar nossa base de vocabulário, vamos tirar vantagem das operações de [dicionários](https://docs.python.org/pt-br/3.13/tutorial/datastructures.html#dictionaries) em Python, a fim de formar nosso **dicionário de vocabulário**, onde cada token receberá um valor inteiro:

````python
lemmatizer = WordNetLemmatizer()

def processing(token):

    '''
    processing:
     - Converte para minúsculo;
     - Remove pontuação;
     - Lematiza.
    '''

    token = token.lower()

    token = re.sub(r'[^\w]', '', token)

    return lemmatizer.lemmatize(token)

````

````python
vocab_dict = {} # definindo o dicionário

for text in token_list:
    for token in text:

        token = processing(token)

        if token not in vocab_dict:
            vocab_dict[token] = len(vocab_dict) + 1

vocab_dict
````

## __4. Indexação de frases:__
<font size=3>

Em posse do dicionário de vocabulário, agora podemos transformar todo _dataset_ em uma representação numérica, onde cada _token_ será indexado de acordo com $\texttt{vocab\_dict}$.

````python
token_ids = []

for text in token_list:
    vec = []

    for token in text:
        token = processing(token)

        vec.append(vocab_dict.get(token))

    token_ids.append(vec)

token_ids
````
