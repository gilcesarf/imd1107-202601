````python
import nltk

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

# Inicializa o lematizador
lemmatizer = WordNetLemmatizer()

from nltk.corpus import stopwords
nltk.download('stopwords')
# Inicializa o conjunto de stop words
englishStopWords = set(stopwords.words('english'))

````

````python

import re
emoji_pattern = re.compile("["
             u"\U0001F600-\U0001F64F"  # emoticons
             u"\U0001F300-\U0001F5FF"  # símbolos & pictogramas
             u"\U0001F680-\U0001F6FF"  # transporte & mapas
             u"\U0001F1E0-\U0001F1FF"  # bandeiras
             u"\U00002700-\U000027BF"  # dingbats
             u"\U0001F900-\U0001F9FF"  # suplemento de símbolos
             u"\U0001FA70-\U0001FAFF"  # símbolos adicionais
             "]+", flags=re.UNICODE)
    
def remove_emojis(text):
  return emoji_pattern.sub(r'', text)

def tweet_preprocess(tweet):
  tweet = tweet.lower()
  tweet = remove_emojis(tweet)
  tweet = re.sub(r'https?://\S+|www\.\S+', '', tweet) #remove links
  tweet = re.sub(r'#\w+', '', tweet) #remove hashtags
  tweet = re.sub(r'\d+', '', tweet) #remove dígitos
  tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet) #remove pontuações
  tweet = re.sub(r'\s+', ' ', tweet) #remove espaços em excesso
  tweet = tweet.strip() #remove espaços em excesso
  new_text = []
  for token in tweet.split(" "): # divide o tweet em tokens
    token_processed = lemmatizer.lemmatize(token) # lematiza o token
    # remove stopwords
    if token_processed not in englishStopWords:
      new_text.append(token_processed) # adiciona o token processado à lista

  tweet = " ".join(new_text) # reconstroi o tweet com os tokens processados
  
  return tweet

lista_tweets_preprocessed = [tweet_preprocess(tweet) for tweet in lista_tweets]

print(f"Primeiro tweet antes e depois do pré-processamento:")
print()
for tweet, original_tweet in zip(lista_tweets_preprocessed[:1], lista_tweets[:1]):
  print(f"original tweet....: {original_tweet}")
  print(f"preprocessed tweet: {tweet}")
  print()
````

### __3. Vetorização:__
<font size=3>

Faça a vetorização TF-IDF dos textos de *tweets* utilizando a classe [`TfidfVectorizar`](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html) do $\texttt{sklearn}$. Para isso:
 1. Utilize o argumento `preprocessor` da classe `TfidfVectorizer` para considerar sua função, desenvolvida na **questão 1**, no pré-processamento;
 2. Faça o *fit* e transformação dos textos;
 3. Imprima na tela os 10 primeiros tokens da lista de vocabulário e o tamanho do vocabulário:
    - Aqui você irá notar alguns tokens estranho e sem significado.

 4. Para filtrar tokens indesejados, utilize os argumentos `min_df` e `max_df` da classe `TfidfVectorizer`:
    * `min_df = x`:
      - Se `x` é inteiro → termo deve aparecer em pelo menos **x documentos**.
      - Se `x` é float → termo deve aparecer em pelo menos **x × 100% dos documentos**.
    
    * `max_df = y`: 
      - Se `y` é inteiro → termo é removido se aparecer em mais de **y documentos**.
      - Se `y` é float → termo é removido se aparecer em mais de **y × 100% dos documentos**.

 6. Quando a lista de vocabulário for satisfatória, **imprima na tela** o primeiro *tweet* vetorizado.


````python
from sklearn.feature_extraction.text import TfidfVectorizer
#limpa stopwords em ingles do TfidfVectorizer, além do nltk aplicado na função tweet_preprocess
tfidf_vectorizer = TfidfVectorizer(
    max_df=0.95,
    min_df=4,
    stop_words="english",
    preprocessor=tweet_preprocess
)

tfidf_vectorizer.fit_transform(lista_tweets)

print(f"Tamanho do vocabulário: {len(tfidf_vectorizer.get_feature_names_out())}")

print(f"Inicio do Vocabulário (10 primeiros tokens):")
print(f"{tfidf_vectorizer.get_feature_names_out()[:10]}")

````

````python
import numpy as np

np.set_printoptions(threshold=np.inf)

tfidf_vectorizer.transform(lista_tweets_preprocessed[:1])

print(f"Primeiro tweet vetorizado: {tfidf_vectorizer.transform(lista_tweets_preprocessed[:1]).toarray()}")
````

### Melhorando a saida do tweet vetorizado

Como a saida do tweet vetorizado não permite uma visualização e uma compreensão adequadas, vamos exibir o tweet vetorizado em um formato mais inteligível

````python
# Posições não nulas do primeiro tweet (token -> valor TF-IDF)
vec = tfidf_vectorizer.transform(lista_tweets_preprocessed[:1])
row = vec[0]

print(f"Primeiro tweet preprocessado:")
print(f"{lista_tweets_preprocessed[:1]}")

print(f"\nPrimeiro tweet vetorizado(coordenadas nao nulas):")
print(f"{row}")

nombres = tfidf_vectorizer.get_feature_names_out()
## pareando tokens e vetorizacao diferente de zero e imprimindo o TFIDF de cada token
print(f"\nTFIDF de cada token:")
for idx, val in zip(row.nonzero()[1], row.data):
    print(f"{nombres[idx]}: {val:.4f}")
````
