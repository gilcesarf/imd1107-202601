__No notebook da aula, não consigo acessar o caminho de um arquivo no drive da disciplina! Como faço?!__

Ao acessar o _link_ compartilhado do drive, o diretório da disciplina "__imd0000-nome_da_disciplina__" ficará no seu drive em "__Compartilhados comigo__" ("Shared with me").

Então faça o seguinte:
 - Entre em "__Compartilhados comigo__";
 - Identifique o diretório da disciplina e clique nos três pontinhos verticais, ao lado do nome;
 - Clique em "__Organizar__" -> "__Adicionar atalho__";
 - Depois clique em "__Todos os locais__" -> "__Meu Drive__" -> "__Adicionar__".

Agora, o diretório da disciplina também estará _sincronizado_ em "__Meu Drive__". Então, ao rodar
